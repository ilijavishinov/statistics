Consider the Bernoulli(p) observations 0 1 0 1 0 0 0 0 0 0
Plot the posterior for p using these priors: Beta(1/2,1/2), Beta(1,1), Beta(10,10), Beta(100,100). 

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
prior_beta_dist_parameters = [
    (1/2, 1/2),
    (1, 1),
    (10, 10),
    (100, 100)
]
x_axis = np.linspace(0,1,500)
for alpha_param, beta_param in prior_beta_dist_parameters:
    plt.plot(x_axis, beta.pdf(x_axis, alpha_param + 2, beta_param + 8),
             label = f'Beta({alpha_param}, {beta_param}) prior')
plt.legend()
```
